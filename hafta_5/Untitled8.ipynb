```python
def convert_rgb_to_gray_level_01(im_1):
    m=im_1.shape[0]
    n=im_1.shape[1]
    im_2=np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            im_2[i,j]=get_distance(im_1[i,j,:])
        return im_2
def get_distance(v,w=[1/3,1/3,1/3]):
    a,b,c=v[0],v[1],v[2]
    w1,w2,w3=w[0],w[1],w[2]
    d=((a**2)*w1+(b**2)*w2+(c**2)*w3)**.5
    return d
```

```python
def get_value_from_triple(temp_1):                  #Send 3 values and take 1 value. We use average in this function.  
    return int(temp_1[0]/3+temp_1[1]/3+temp_1[2]/3)
def convert_rgb_to_gray(image_1):#I wander on my image. Return value from the function(get_value_from_triple) is i and j in the new image.
    m,n,k=image_1.shape
    new_image=np.zeros((m,n),dtype='uint8')  #We create a new image.
    for i in range(m):
        for j in range(n):
            s=get_value_from_triple(image_1[i,j,:])
            new_image[i,j]=s
    return new_image
```

```python
#overflow olmasını engelleyen bir fonksiyon yazdık.
#my_f_1 intensity'i arttırıyor.
def my_f_1(a,b):#a eğer intensity ise "a+5" demek bu değeri 5 arttırmak demek.
    assert a>=0;"intensity pozitive","error intensity not pozitive"
    if(a<=255-b):
           return a+b
    else:
        return 255
my_f_1(244,30)  
```

```python
#my_f_2 intensity'nin complement'ini alıyor. 
def my_f_2(a):#a eğer intensity ise "a+5" demek bu değeri 5 arttırmak demek.
    #assert a>=0;"intensity pozitive","error intensity not pozitive"
    return int(255-a)
my_f_2(243)  
```
