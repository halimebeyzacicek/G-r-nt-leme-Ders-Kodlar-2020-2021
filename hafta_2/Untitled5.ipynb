```python
def display_two_image(im_1,im_2):
    plt.subplot(1,2,1)
    plt.imshow(image_1)

    plt.subplot(1,2,2)
    plt.imshow(image_2+30)

    plt.show
def rotate(im_1):
    m,n,k=im_1.shape
    new_image=np.zeros((n,m,k),dtype='uint8')
    
    for i in range (m):
        for j in range (n):
            temp=image_1[i,j]
            new_image[j,i]=temp
    return new_image
     
```

```python
image_1=plt.imread('pic_2.jpg')
image_2=rotate(image_1)
display_two_image(image_1,image_2)
```
