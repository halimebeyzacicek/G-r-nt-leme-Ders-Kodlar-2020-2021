```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
new_m=int(m/2) #new sizes
new_n=int(n/2)
new_m,new_n
```

```python
def my_small_image(img_f): #this function makes reduce the size.
    m,n,p=img_f.shape
    
    new_m=int(m/2) #new sizes
    new_n=int(n/2)
    img_small=np.zeros((new_m,new_n,3),dtype=int) #new image
    
    for m in range(new_m):
        for n in range(new_n):
            img_small[m,n,0]=img_f[m*2,n*2,0]
            img_small[m,n,1]=img_f[m*2,n*2,1]
            img_small[m,n,2]=img_f[m*2,n*2,2]
    return img_small        
```

```python
img_new=my_small_image(img_1)
img_new_1=my_small_image(img_new)
img_new_2=my_small_image(img_new_1)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(2,2,1),plt.imshow(img_1) #%100--------y=1000
plt.subplot(2,2,2),plt.imshow(img_new) #%50-------y=500
plt.subplot(2,2,3),plt.imshow(img_new_1) #%25-----y=250
plt.subplot(2,2,4),plt.imshow(img_new_2) #%12,5---y=125
```
