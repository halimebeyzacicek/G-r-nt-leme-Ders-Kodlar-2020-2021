```python
import numpy as np 
import matplotlib.pyplot as plt
image_1=plt.imread(jpg_files[2])
```

```python
def get_value_from_triple(temp_1):        
    return int(temp_1[0]/3+temp_1[1]/3+temp_1[2]/3)
def get_0_1_value_from_triple(temp_1):  #3 sayıyı temsilen bir syaıyı bulmaktır.
    #temp_1=image_1[0,0,:]              #en iyi yöntem üçünün oratalamasını almaktır.
    temp=int(temp_1[0]/3+temp_1[1]/3+temp_1[2]/3)
    if temp<110:
        return 0
    else:
        return 1
```

```python
def convert_rgb_to_bw(image_1):
    m,n,k=image_1.shape
    new_image=np.zeros((m,n),dtype='uint8')
    for i in range(m):
        for j in range(n):
            s=get_0_1_value_from_triple(image_1[i,j,:])
            new_image[i,j]=s
    return new_image
def convert_rgb_to_gray(image_1):
    m,n,k=image_1.shape
    new_image=np.zeros((m,n),dtype='uint8')
    for i in range(m):
        for j in range(n):
            s=get_value_from_triple(image_1[i,j,:])
            new_image[i,j]=s
    return new_image
```

```python
im_1_gray=convert_rgb_to_gray(image_1)
im_1_bw=convert_rgb_to_bw(image_1)
```

```python
plt.subplot(1,3,1)
plt.imshow(image_1)

plt.subplot(1,3,2)
plt.imshow(im_1_gray,cmap='gray')

plt.subplot(1,3,3)
plt.imshow(im_1_bw,cmap='gray')

plt.show()
```
